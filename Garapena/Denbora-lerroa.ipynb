```
with open('EhuHac/hasierak.txt') as f:
    lerroak = f.read().split('\n')

lerroak = lerroak[:136]
lerroak.remove('')
zenbakiak = [int(lerroa) for lerroa in lerroak]
```

```
hutsak_es = []
with open('EhuHac/EhuHac-es.txt') as f:
    for i, lerroa in enumerate(f, 1):
        if lerroa == '\n':
            hutsak_es.append(i)

hutsak_en = []
with open('EhuHac/EhuHac-en.txt') as f:
    for i, lerroa in enumerate(f, 1):
        if lerroa == '\n':
            hutsak_en.append(i)

hutsak_eu = []
with open('EhuHac/EhuHac-eu.txt') as f:
    for i, lerroa in enumerate(f, 1):
        if lerroa == '\n':
            hutsak_eu.append(i)
```

```
import matplotlib.pyplot as plt
```
