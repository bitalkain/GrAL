```python
def zuzen_ordenatuta(zenbakiak):
    azkena = 0
    for multzoa in zenbakiak:
        for zenbakia in multzoa:
            if zenbakia <= azkena:
                print(zenbakia, azkena)
                return False
            azkena = zenbakia
    return True
```

```python
def elkartu_filma(es_zenb, en_zenb, eseu_esf, eseu_euf, eneu_enf, eneu_euf):
    #if zuzen_ordenatuta(es_zenb) and zuzen_ordenatuta(en_zenb):
    es_ix = 1
    en_ix = 1
    gordeak_es = []
    gordeak_en = []
    gordeak_eu = []
    uneko_zenb_es = es_zenb[0].copy()
    uneko_zenb_en = en_zenb[0].copy()
    uneko_str_es = eseu_esf[0]
    uneko_str_en = eneu_enf[0]
    uneko_str_eseu = eseu_euf[0]
    uneko_str_eneu = eneu_euf[0]
    azkena = ''
    arazo_kop = 0
    while es_ix < len(es_zenb) and en_ix < len(en_zenb):
        
        if uneko_zenb_es == uneko_zenb_en:
            if uneko_str_eseu == uneko_str_eneu:
                gordeak_es.append(uneko_str_es)
                gordeak_en.append(uneko_str_en)
                gordeak_eu.append(uneko_str_eneu)
                uneko_zenb_es = es_zenb[es_ix].copy()
                uneko_zenb_en = en_zenb[en_ix].copy()
                uneko_str_es = eseu_esf[es_ix]
                uneko_str_en = eneu_enf[en_ix]
                uneko_str_eseu = eseu_euf[es_ix]
                uneko_str_eneu = eneu_euf[en_ix]
                es_ix += 1
                en_ix += 1
            else:
                if arazo_kop < 10:
                    print('Arazoa')
                    print(uneko_zenb_es, uneko_zenb_en, 
                            uneko_str_eseu, uneko_str_eneu)
                arazo_kop += 1
                uneko_zenb_es = es_zenb[es_ix].copy()
                uneko_zenb_en = en_zenb[en_ix].copy()
                uneko_str_es = eseu_esf[es_ix]
                uneko_str_en = eneu_enf[en_ix]
                uneko_str_eseu = eseu_euf[es_ix]
                uneko_str_eneu = eneu_euf[en_ix]
                es_ix += 1
                en_ix += 1

        elif uneko_zenb_es[0] < uneko_zenb_en[0]:
            uneko_zenb_es = es_zenb[es_ix].copy()
            uneko_str_es = eseu_esf[es_ix]
            uneko_str_eseu = eseu_euf[es_ix]
            es_ix += 1
        elif uneko_zenb_en[0] < uneko_zenb_es[0]:
            uneko_zenb_en = en_zenb[en_ix].copy()
            uneko_str_en = eneu_enf[en_ix]
            uneko_str_eneu = eneu_euf[en_ix]
            en_ix += 1
        
        elif uneko_zenb_es == uneko_zenb_en[:len(uneko_zenb_es)]:
            uneko_zenb_es.extend(es_zenb[es_ix])
            uneko_str_es += ' ' + eseu_esf[es_ix]
            uneko_str_eseu += ' ' + eseu_euf[es_ix]
            es_ix += 1
            azkena = 'es'
        elif uneko_zenb_en == uneko_zenb_es[:len(uneko_zenb_en)]:
            uneko_zenb_en.extend(en_zenb[en_ix])
            uneko_str_en += ' ' + eneu_enf[en_ix]
            uneko_str_eneu += ' ' + eneu_euf[en_ix]
            en_ix += 1
            azkena = 'en'

        else:
            if azkena == 'es':
                uneko_zenb_es = es_zenb[es_ix - 1].copy()
                uneko_zenb_en = en_zenb[en_ix].copy()
                uneko_str_es = eseu_esf[es_ix - 1]
                uneko_str_en = eneu_enf[en_ix]
                uneko_str_eseu = eseu_euf[es_ix - 1]
                uneko_str_eneu = eneu_euf[en_ix]
                en_ix += 1
            elif azkena == 'en':
                uneko_zenb_es = es_zenb[es_ix].copy()
                uneko_zenb_en = en_zenb[en_ix - 1].copy()
                uneko_str_es = eseu_esf[es_ix]
                uneko_str_en = eneu_enf[en_ix - 1]
                uneko_str_eseu = eseu_euf[es_ix]
                uneko_str_eneu = eneu_euf[en_ix - 1]
                es_ix += 1
            else:
                print('Exekuzioak ez zuen hona iritsi behar...')
                return

    if uneko_zenb_es == uneko_zenb_en:
        if uneko_str_eseu == uneko_str_eneu:
            gordeak_es.append(uneko_str_es)
            gordeak_en.append(uneko_str_en)
            gordeak_eu.append(uneko_str_eneu)
        else:
            print('Arazoa')
            print(uneko_zenb_es, uneko_zenb_en, 
                        uneko_str_eseu, uneko_str_eneu)
            #return

    return gordeak_es, gordeak_en, gordeak_eu

    #print('Ez daude ondo ordenatuta')
```

```python
def elkartu_filma_berezia(es_zenb, en_zenb, 
                          eseu_esf, eseu_euf, eneu_enf, eneu_euf):
    #if zuzen_ordenatuta(es_zenb) and zuzen_ordenatuta(en_zenb):
    es_ix = 1
    en_ix = 1
    gordeak_es = []
    gordeak_en = []
    gordeak_eseu = []
    gordeak_eneu = []
    uneko_zenb_es = es_zenb[0].copy()
    uneko_zenb_en = en_zenb[0].copy()
    uneko_str_es = eseu_esf[0]
    uneko_str_en = eneu_enf[0]
    uneko_str_eseu = eseu_euf[0]
    uneko_str_eneu = eneu_euf[0]
    azkena = ''
    while es_ix < len(es_zenb) and en_ix < len(en_zenb):
        
        if uneko_zenb_es == uneko_zenb_en:
            gordeak_es.append(uneko_str_es)
            gordeak_en.append(uneko_str_en)
            gordeak_eseu.append(uneko_str_eseu)
            gordeak_eneu.append(uneko_str_eneu)
            uneko_zenb_es = es_zenb[es_ix].copy()
            uneko_zenb_en = en_zenb[en_ix].copy()
            uneko_str_es = eseu_esf[es_ix]
            uneko_str_en = eneu_enf[en_ix]
            uneko_str_eseu = eseu_euf[es_ix]
            uneko_str_eneu = eneu_euf[en_ix]
            es_ix += 1
            en_ix += 1

        elif uneko_zenb_es[0] < uneko_zenb_en[0]:
            uneko_zenb_es = es_zenb[es_ix].copy()
            uneko_str_es = eseu_esf[es_ix]
            uneko_str_eseu = eseu_euf[es_ix]
            es_ix += 1
        elif uneko_zenb_en[0] < uneko_zenb_es[0]:
            uneko_zenb_en = en_zenb[en_ix].copy()
            uneko_str_en = eneu_enf[en_ix]
            uneko_str_eneu = eneu_euf[en_ix]
            en_ix += 1
        
        elif uneko_zenb_es == uneko_zenb_en[:len(uneko_zenb_es)]:
            uneko_zenb_es.extend(es_zenb[es_ix])
            uneko_str_es += ' ' + eseu_esf[es_ix]
            uneko_str_eseu += ' ' + eseu_euf[es_ix]
            es_ix += 1
            azkena = 'es'
        elif uneko_zenb_en == uneko_zenb_es[:len(uneko_zenb_en)]:
            uneko_zenb_en.extend(en_zenb[en_ix])
            uneko_str_en += ' ' + eneu_enf[en_ix]
            uneko_str_eneu += ' ' + eneu_euf[en_ix]
            en_ix += 1
            azkena = 'en'

        else:
            if azkena == 'es':
                uneko_zenb_es = es_zenb[es_ix - 1].copy()
                uneko_zenb_en = en_zenb[en_ix].copy()
                uneko_str_es = eseu_esf[es_ix - 1]
                uneko_str_en = eneu_enf[en_ix]
                uneko_str_eseu = eseu_euf[es_ix - 1]
                uneko_str_eneu = eneu_euf[en_ix]
                en_ix += 1
            elif azkena == 'en':
                uneko_zenb_es = es_zenb[es_ix].copy()
                uneko_zenb_en = en_zenb[en_ix - 1].copy()
                uneko_str_es = eseu_esf[es_ix]
                uneko_str_en = eneu_enf[en_ix - 1]
                uneko_str_eseu = eseu_euf[es_ix]
                uneko_str_eneu = eneu_euf[en_ix - 1]
                es_ix += 1
            else:
                print('Exekuzioak ez zuen hona iritsi behar...')
                return

    if uneko_zenb_es == uneko_zenb_en:
        gordeak_es.append(uneko_str_es)
        gordeak_en.append(uneko_str_en)
        gordeak_eseu.append(uneko_str_eseu)
        gordeak_eneu.append(uneko_str_eneu)


    return gordeak_es, gordeak_en, gordeak_eseu, gordeak_eneu

    #print('Ez daude ondo ordenatuta')
```
