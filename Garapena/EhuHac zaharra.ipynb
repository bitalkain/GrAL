```
with open('EhuHac.es-eu.eu') as f1, open('EhuHac.en-eu.eu') as f2:
    aurkitua = True
    esaldiak = []
    while aurkitua:
        aurkitua = False
        lerroa1 = f1.readline()
        lerroa2 = f2.readline()
        if not lerroa1 or not lerroa2:
            break
        if lerroa1 == lerroa2:
            #print(lerroa1)
            esaldiak.append(lerroa1)
            aurkitua = True
        else:
            #has_pos1 = f1.tell()
            has_pos2 = f2.tell()
            for _ in range(5):
                if lerroa1 == f2.readline():
                    #print(lerroa1)
                    esaldiak.append(lerroa1)
                    aurkitua = True
                    break
            if not aurkitua:
                f2.seek(has_pos2)
                for _ in range(5):
                    if lerroa2 == f1.readline():
                        #print(lerroa2)
                        esaldiak.append(lerroa2)
                        aurkitua = True
                        break
```

```
link = 'https://www.ehu.eus/ehg/hac/liburua'
f = urlopen(link)
liburuak_html = f.read()
root = lxml.html.fromstring(liburuak_html)
elems = root.cssselect('span.tituloak')
liburu_linkak = [elem.getchildren()[0].attrib['href'] for elem in elems]
```

```
liburu_linkak.pop(-6) # Voltairen esteka ez dabil
```

```
or_kopuruak = []
for lib_link in liburu_linkak:
    print(lib_link)
    link = 'https://www.ehu.eus/ehg/hac/'+lib_link
    f = urlopen(link)
    orria1_html = f.read()
    root = lxml.html.fromstring(orria1_html)
    elems = root.cssselect('div.nabi')
    or_kop = elems[0].text_content().split()[0]
    or_kopuruak.append(int(or_kop))
```

```
or_kopuruak
```

```
esaldiak = []
frases = []
sentences = []
for lib_link, or_kop in zip(liburu_linkak, or_kopuruak):
    for orria in range(1, or_kop+1):
        link = 'https://www.ehu.eus/ehg/hac/'+lib_link+'&o='+str(orria)
        print(link)
        f = urlopen(link)
        orria_html = f.read()
        root = lxml.html.fromstring(orria_html)
        elems = root.cssselect('div.esalBi')
        if len(elems) % 4 != 0:
            print("Arazoa")
        else:
            esaldiak = []
            frases = []
            sentences = []
            for i in range(0, len(elems), 4):
                esaldiak.append(elems[i].text_content())
                frases.append(elems[i+1].text_content())
                sentences.append(elems[i+3].text_content())
```

```
with open('EhuHac-eu.txt', 'w') as f:
    for esaldia in esaldiak:
        f.write(esaldia+'\n')

with open('EhuHac-en.txt', 'w') as f:
    for sentence in sentences:
        f.write(sentence+'\n')

with open('EhuHac-es.txt', 'w') as f:
    for frase in frases:
        f.write(frase+'\n')
```
