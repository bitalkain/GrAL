```
with open('EhuHac/EhuHac-trainorig-es.txt') as f1,\
     open('EhuHac/EhuHac-trainorig-en.txt') as f2,\
     open('EhuHac/EhuHac-trainorig-eu.txt') as f3:
    badaudenak = set()
    for lerroa_es, lerroa_en, lerroa_eu in zip(f1, f2, f3):
        badaudenak.add((lerroa_es[:-1], lerroa_en[:-1], lerroa_eu[:-1]))
```

```
len(badaudenak)
```

# Biblia gabekoei balidaziokoak kendu

```
gordetzeko_es = []
gordetzeko_en = []
gordetzeko_eu = []

with open('EhuHac/EhuHac-2hbibliakenduta-es.txt') as f1,\
     open('EhuHac/EhuHac-2hbibliakenduta-en.txt') as f2,\
     open('EhuHac/EhuHac-2hbibliakenduta-eu.txt') as f3:
    for lerroa_es, lerroa_en, lerroa_eu in zip(f1, f2, f3):
        if (lerroa_es[:-1], lerroa_en[:-1], lerroa_eu[:-1]) in badaudenak:
            gordetzeko_es.append(lerroa_es)
            gordetzeko_en.append(lerroa_en)
            gordetzeko_eu.append(lerroa_eu)
```

```
len(gordetzeko_es), len(gordetzeko_en), len(gordetzeko_eu)
```

```
with open('EhuHac/EhuHac-trainorigbibgabe-es.txt', 'w') as f:
    f.write(''.join(gordetzeko_es))

with open('EhuHac/EhuHac-trainorigbibgabe-en.txt', 'w') as f:
    f.write(''.join(gordetzeko_en))

with open('EhuHac/EhuHac-trainorigbibgabe-eu.txt', 'w') as f:
    f.write(''.join(gordetzeko_eu))
```
